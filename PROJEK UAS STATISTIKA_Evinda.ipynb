Menentukan batas atas dan batas bawah yang akan digunakan untuk mengganti nilai outliers dengan batas atas dan bawah

```python
def batas_atas_bawah(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR

    print("Nilai Batas Bawah = ", batas_bawah)
    print("Nilai Batas Atas = ", batas_atas)

print("Penduduk Miskin")
batas_atas_bawah(data['Penduduk Miskin'])

print("\nTingkat Pengagguran Terbuka")
batas_atas_bawah(data['TPT'])

print("\nIndeks Pembangunan Manusia")
batas_atas_bawah(data['IPM'])

print("\nBANSOS")
batas_atas_bawah(data['BANSOS'])

print("\nGini Rasio")
batas_atas_bawah(data['GINI RASIO'])
```

Mengetahui nilai outliers pada setiap variabel

```python
#Menampilkan nilai outliers dari setiap variabel
def cek_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR

    outliers = (data < batas_bawah) | (data > batas_atas)
    return outliers

outliers = cek_outliers(data['Penduduk Miskin'])
nilai_outlier = data[outliers]
print("Nilai Outlier Penduduk Miskin:")
print(nilai_outlier[['Kabupaten/Kota Se Jawa Timur', 'Penduduk Miskin']])

outliers = cek_outliers(data['TPT'])
nilai_outlier = data[outliers]
print("\nNilai Outlier TPT:")
print(nilai_outlier[['Kabupaten/Kota Se Jawa Timur', 'TPT']])

outliers = cek_outliers(data['IPM'])
nilai_outlier = data[outliers]
print("\nNilai Outlier IPM:")
print(nilai_outlier[['Kabupaten/Kota Se Jawa Timur', 'IPM']])

outliers = cek_outliers(data['BANSOS'])
nilai_outlier = data[outliers]
print("\nNilai Outlier BANSOS:")
print(nilai_outlier[['Kabupaten/Kota Se Jawa Timur', 'BANSOS']])

outliers = cek_outliers(data['GINI RASIO'])
nilai_outlier = data[outliers]
print("\nNilai Outlier GINI RASIO:")
print(nilai_outlier[['Kabupaten/Kota Se Jawa Timur', 'GINI RASIO']])
```

<p1>Mengganti nilai outliers dengan nilai batas bawah jika nilai outliers berada dibawah batas bawah dan mengganti nilai batas atas jika nilai outliers berada diatas batas atas</p1>

```python
data_baru = data.copy()
def replace_outliers(data_baru):
    # Identifikasi outliers
    Q1 = data_baru.quantile(0.25)
    Q3 = data_baru.quantile(0.75)
    IQR = Q3 - Q1

    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    outliers_bawah = data_baru < batas_bawah
    outliers_atas = data_baru > batas_atas

    # Ganti nilai outliers dengan batas bawah atau batas atas
    data_baru[outliers_bawah] = batas_bawah
    data_baru[outliers_atas] = batas_atas

    return data_baru

data_baru['BANSOS'] = replace_outliers(data_baru['BANSOS'])
```
